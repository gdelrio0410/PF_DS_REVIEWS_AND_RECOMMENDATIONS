```python
import pandas as pd
import json
import pickle
import seaborn as sns

```

```python

def abrir_json_y_crear_pandas(ruta_archivo):
    data = []
    with open(ruta_archivo, 'r',encoding='utf-8') as archivo:
        for linea in archivo:
            try:
                objeto = json.loads(linea)
                data.append(objeto)
            except json.JSONDecodeError as e:
                print(f"Error al cargar el JSON en la línea {e.lineno}: {e.msg}")

    df = pd.DataFrame(data)
    
    return df
```

```python
data2=pd.read_csv("../05 - Yat/data/rew_modi.csv")
```

```python
# Realizar el groupby, sumar los valores y contar las repeticiones
data2_promedio= data2_promedio.groupby('business_id').agg({'stars': ['sum', 'count']})

# Renombrar las columnas resultantes
data2_promedio.columns = ['Suma', 'Rewiews']

# Mostrar el resultado

data2_promedio['Promedio'] = data2_promedio['Suma'] / data2_promedio['Rewiews']
data2_promedio['Promedio'] =data2_promedio['Promedio'].round(2)
data2_promedio=data2_promedio.reset_index()
data2_promedio=data2_promedio[['business_id','Rewiews','Promedio']]
data2_promedio


```

```python
def filtrar_dataframe(dataframe, columna, frase):
    nuevo_dataframe = dataframe[dataframe[columna].str.contains(frase)]
    return nuevo_dataframe
```

# Cantidad de comentarios

```python
df=data_comple[data_comple['state'] == "TN"]
Categorias=pd.DataFrame(df[['name',"state","latitude","longitude","Rewiews","Promedio"]])
#sns.barplot(y='category', x='count', data=Categorias)

Categorias.sort_values('Rewiews', ascending=False, inplace=True)
Categorias.reset_index(drop=True, inplace=True)
Ubicacion=Categorias.head(5)
result=Categorias[["name","state","Rewiews","Promedio"]]
result
result["Promedio"] = result.groupby("name")["Promedio"].transform('mean')

# Eliminar los duplicados en la columna 'A'
result= result.drop_duplicates(subset="name")


result.to_csv("../01 - Data_tratados/Puntuacion.csv")
```
