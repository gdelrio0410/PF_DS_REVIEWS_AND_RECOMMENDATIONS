### Base de Datos Población EEUU

```python
import pandas as pd
```

```python
#Leemos el csv de Poblacion de EEUU 
df = pd.read_csv('Dataset/DataSet Population EEUU.csv')

df
```

```python
df = df.drop(columns=['SEX AND AGE'])
```

```python
nuevos_nombres = {col: col.replace("SEX AND AGE", "") for col in df.columns}

# Renombra las columnas del DataFrame usando el diccionario
df.rename(columns=nuevos_nombres, inplace=True)
nuevos_nombres = {col: col.replace("!!", "") for col in df.columns}

# Renombra las columnas del DataFrame usando el diccionario
df.rename(columns=nuevos_nombres, inplace=True)
```

```python
df
```

```python
# Aquí seleccionamos todas las columnas hasta la 5ta.
nombres_columnas = df.columns.tolist() 
df = df.drop(df.columns[5:18], axis=1)
```

```python
df = df.drop(df.columns[10:], axis=1)
```

```python
df
```

```python
df.iloc[:, 1:] = df.iloc[:, 1:].apply(lambda x: x.shift(-1))
df = df.iloc[::3]
df = df.reset_index(drop=True)
```

```python
df
```

```python
df.rename(columns={'Label (Grouping)': 'Estado'}, inplace=True)
```

```python
df
```

```python
df.info()
```

```python
for x in range(0, df.shape[1]):
    print(f"La columna {x} es de tipo de dato: {type(df.iloc[0,x])}")
```

```python
def eliminar_comas(cadena):
    return cadena.replace(',', '')

# Aplicar la función a todas las columnas del DataFrame
df = df.applymap(eliminar_comas)
```

```python
df
```

```python
#convertir todas las columnas desde la segunda columna hasta el final a tipo float
columnas_a_convertir = df.columns[1:]
# Aplicar la conversión a tipo entero a las columnas seleccionadas
for columna in columnas_a_convertir:
    df[columna] = df[columna].astype(float)
```

```python
df.to_csv('Population', index= False)   
```
